```python
import polars as pl
from datasets import load_dataset
import os
```

```python
print("Loading EDNet KT4 (Research-grade slice) in streaming mode...")

try:
    dataset_stream = load_dataset("mgor/EDNet", "kt4", split="train", streaming=True)
    
    print(" Connection successful! Fetching first record...")
    first_record = next(iter(dataset_stream))
    print("\nSample Data Point:")
    print(first_record)

except Exception as e:
    print(f" Error: {e}")

print("\n Converting sample to Polars for inspection...")
sample_size = 50000
sample_data = list(dataset_stream.take(sample_size))
df = pl.DataFrame(sample_data)

print(f" Preview of {sample_size} interactions:")
print(df.head())
```
